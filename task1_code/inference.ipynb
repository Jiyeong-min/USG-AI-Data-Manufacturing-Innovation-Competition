```python
import os
import pandas as pd
from PIL import Image
from transformers import ViTForImageClassification, ViTImageProcessor, pipeline
from tqdm.auto import notebook_tqdm
```

```python
model_name = "google/vit-base-patch16-224"
processor = ViTImageProcessor.from_pretrained(model_name, return_tensor="pt")
```

```python
model = ViTForImageClassification.from_pretrained("./results/vit-experience-1/checkpoint-866/")
model.eval()

classifier = pipeline("image-classification", model=model, image_processor=processor, device="cuda")
```

```python
sub_df = pd.read_csv("./task1/answer_sample.csv")
sub_df.head()
```

```python
image_list = sorted(os.listdir("./task1/test/"))
image_list = [f for f in image_list if f.endswith('.jpeg')]
```

```python
predict_labels = []
for image_name in image_list:
    image = Image.open(os.path.join("./task1/test/", image_name))
    predict_labels.append(classifier(image)[0]['label'])
```

```python
predict_labels = [0 if x == 'good' else 1 for x in predict_labels]
predict_labels
```

```python
sub_df['bad'] = predict_labels
sub_df.to_csv('./submission.csv', index=False)
```
